```python
input_data = '/Volumes/ANNA_HD/ANALYSIS/DLC_projects/2025/seabird-marc-2025-11-06/videos_v1/FILE00009_sDLC_DekrW32_seabirdNov6shuffle1_snapshot_170_el_filtered.h5'
individuals = ['bird1', 'bird2', 'bird3', 'bird4', 'bird5']
bodyparts = ['leftwing', 'rightwing', 'middle']
interpolator = Interpolation(threshold=0.60, interpolation_method='linear')
interpolated_data = interpolator.interpolate_data(input_data, bodyparts)

output_dir = '/Users/annateruel/Desktop/'
if isinstance(interpolated_data, dict):
    for filename, df in interpolated_data.items():
        output_path = os.path.join(output_dir, filename)
        df.to_hdf(output_path, key='df', mode='w')
else:
    output_path = os.path.join(output_dir, 'filtered.h5')
    interpolated_data.to_hdf(output_path, key='df', mode='w')

print(f"Interpolated data saved to {output_dir}")
```
